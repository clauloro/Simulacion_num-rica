# Diferencias Finitas - Método de Euler

## Ecuación Diferencial Ordinaria (EDO)

En este notebook, resolvemos la ecuación diferencial ordinaria:

$$
\frac{du}{dt} = u, \quad u(0) = 1, \quad 0 \leq t \leq 4
$$

usando el **método de Euler** con diferentes pasos de tiempo \( \Delta t \).

Compararemos los resultados obtenidos con la **solución analítica**:

$$
u(t) = e^t
$$


## Implementación del Método de Euler

El método de Euler es una técnica numérica basada en diferencias finitas para resolver ecuaciones diferenciales de primer orden.

Se define como:

$$
u_{n+1} = u_n + \Delta t \cdot f(u_n, t_n)
$$

donde \( f(u, t) = u \) en nuestro caso.

Implementamos la función en Python para resolver la ecuación con diferentes valores de \( \Delta t \).


```python
import numpy as np
import matplotlib.pyplot as plt

def metodo_euler(f, u0, t0, tf, dt):
    """
    Método de Euler para resolver u' = f(u, t) con condición inicial u(t0) = u0.
    """
    t_values = np.arange(t0, tf + dt, dt)
    u_values = np.zeros(len(t_values))
    u_values[0] = u0
    
    for i in range(1, len(t_values)):
        u_values[i] = u_values[i - 1] + dt * f(u_values[i - 1], t_values[i - 1])
    
    return t_values, u_values

# Definir la función f(u, t) = u
def f(u, t):
    return u

# Parámetros
t0, tf = 0, 4  # Intervalo de tiempo
u0 = 1         # Condición inicial

# Valores de dt
dt_values = [0.04, 0.2]

# Solución analítica
t_exact = np.linspace(t0, tf, 100)
u_exacta = np.exp(t_exact)

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(t_exact, u_exacta, 'b-', label='Solución analítica')

for dt in dt_values:
    t_values, u_values = metodo_euler(f, u0, t0, tf, dt)
    plt.plot(t_values, u_values, 'o', label=f'dt={dt}')

plt.xlabel('Time')
plt.ylabel('u(t)')
plt.legend()
plt.grid()
plt.title('Método de Euler vs Solución Exacta')
plt.show()
```

## Comparación entre la Solución Numérica y la Solución Analítica

A continuación, graficamos la solución exacta \( u(t) = e^t \) junto con los resultados obtenidos mediante el método de Euler con:

- \(  dt = 0.04 \) (más preciso)
- \(  dt = 0.2 \) (menos preciso)

Observamos cómo la precisión del método de Euler mejora cuando reducimos \( dt \).
